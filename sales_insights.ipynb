merging records

```python
df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")

files = [file for file in os.listdir("./Sales_Data")]

all_months_data = pd.DataFrame()
for file in files :
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])
    
all_months_data.to_csv("all_data.csv", index=False)
```

```python
#finding nan 
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data=all_data.dropna(how='all')
all_data.head()
```

```python
#finding or

temp_df = all_data[all_data['Order Date'].str[0:2] == 'Or']
temp_df.head(50)
```

```python
#finding or

temp_df = all_data[all_data['Order Date'].str[0:2] == 'Or']
temp_df.head(50)
```

```python
all_data= all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data
```

```python
# converting column data type in correct type

all_data['Quantity Ordered']=all_data['Quantity Ordered'].astype('int32')
all_data['Price Each']=all_data['Price Each'].astype('float')
all_data
```

```python
#adding Month column

all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

#adding city column

all_data['City']=all_data['Purchase Address'].apply(lambda x: x.split(',')[1])
all_data.head()

```python
#adding city column
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City']=all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")

all_data.head()
```

best month for sale and how much earned that month

```python
all_data['Sales'] = all_data['Quantity Ordered'].astype('int') * all_data['Price Each'].astype('float')
```

```python
results = all_data.groupby('Month').sum()
results.head()
```

```python
import matplotlib.pyplot as plt

months=range(1,13)
print(months)

plt.bar(months,results['Sales'])
plt.xticks(months)
plt.title('Total Sales',fontsize=17)
plt.ylabel('Sales in USD Millions ($)',fontsize=15)
plt.xlabel('Month Number',fontsize=15)
plt.show()
```

what products are most often sold together?

```python
df=all_data[all_data['Order ID'].duplicated(keep=False)]
df.head(15)
```

```python
df['Grouped']= df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head()
```

```python
df=df[['Order ID','Grouped']].drop_duplicates()
df.head(20)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,3)))
    
for key, value in count.most_common(10):
    print(key, value)
```

what product sold the most and why?

```python
product_gr=all_data.groupby('Product')
#product_gr.sum()
qty_ordered=product_gr.sum()['Quantity Ordered']
products = [product for product , df in product_gr]

plt.bar(products, qty_ordered)
plt.xticks(products, rotation='vertical', size=9)
plt.grid()
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']
#print(prices)
fig, ax1=plt.subplots()

ax2=ax1.twinx()
ax1.bar(products, qty_ordered, color='g')
ax2.plot(products,prices, '-b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('quantity Ordered', color='g')
ax2.set_ylabel('Price $', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)
plt.show()
```
